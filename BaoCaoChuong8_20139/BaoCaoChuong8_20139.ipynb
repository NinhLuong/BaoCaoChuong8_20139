```python
import pandas as pd
from pandas import Series, DataFrame
from datetime import datetime
# tạo một biểu đồ có 1 hàng 1 cột và biểu đồ đang vẽ ở vị trí 1
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# đọc file dữ liệu
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True)
# gán cột SPX của dữ liệu 
spx = data['SPX']
# vẽ dự liệu đã gán ra dạng đường theo thời gian
spx.plot(ax=ax, style='k-')
# tạo một list chưa các tuple là điểm ghi chú tại móc thời gian
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),    
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')]

# dùng vòng for để lấy giá trị tại vị trí ứng trong list cristis_data
for date, label in crisis_data:
    # ứng với date, label trong list dùng method annotate để đăt điểm ghi chú
    # đạt mũi tên chỉ vào các điểm tời gian 50 và đặt text cách 200 
    # với định dạng là màu đen và nằm phía trên bên trái 
    ax.annotate(label, xy=(date, spx.asof(date) + 50),xytext=(date, spx.asof(date) + 200), arrowprops=dict(facecolor='black'), horizontalalignment='left', verticalalignment='top')
# phóng to các giá trị hiện thị trên đồ thi trong khoản 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in 2008-2009 financial crisis') #đặt tiêu đề 
```

```python
# loại bỏ các dữ liệu theo các điều kiện truyền vào
data = data[(data.LATITUDE > 18) & (data.LATITUDE < 20) &
(data.LONGITUDE > -75) & (data.LONGITUDE < -70)
& data.CATEGORY.notnull()]
# tạo hàm để tách và xử lý lấy các dữ liệu chuỗi
# tách các dấu phẩy bên các chuỗi trong dữ liệu 
def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]
# lấy ra từng phần tử sau khi đã tách dấu phẩy và sắp xếp lại 
def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series)
    return sorted(set.union(*cat_sets))
# loại bỏ dấu . và | có trong chuỗi 
def get_english(cat):
    code, names = cat.split('.')
    if '|' in names:
        names = names.split(' | ')[1]
    return code, names.strip()

#thử và chỏa kết quả mong muốn 
get_english('2. Urgences logistiques | Vital Lines')
```

```python
#xử lý chuỗi ký tự tác bỏ dấy (,) và sắp xếp dữ liệu
all_cats = get_all_categories(data.CATEGORY)
# tách bỏ dấu(.) và (|)
english_mapping = dict(get_english(x) for x in all_cats)
english_mapping['2a']
```

```python
# tách bỏ các dấu (.) trong chuổi ký tự 
def get_code(seq):
 return [x.split('.')[0] for x in seq if x]

all_codes = get_code(all_cats)
# trích xuất các mã danh mục duy nhất
code_index = pd.Index(np.unique(all_codes))
# xây dựng một DataFrame gồm các số 0 có các số đó làm cột và cùng chỉ mục với dữ liệu
dummy_frame = DataFrame(np.zeros((len(data), len(code_index))),
index=data.index, columns=code_index)
# hiện thị tất cả các hàng của cột từ 0 đến 6
dummy_frame.iloc[:, :6]

```

```python
dummy_frame
```

```python

for row, cat in zip(data.index, data.CATEGORY):
    codes = get_code(to_cat_list(cat)) # xử lý dấu (,)
    # đặt các mục thích hợp của mỗi hàng thành 1
    dummy_frame.loc[row,codes]=1

#cuối cùng nối mục này với dữ liệu
data=data.join(dummy_frame.add_prefix('category_')) 
data.iloc[:, 10:15]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap 
def basic_haiti_map(ax=None, lllat=17.25, urlat=20.25,
    lllon=-75, urlon=-71):
    # tạo lập thể cực Basemap 
    m = Basemap(ax=ax, projection='stere',
        lon_0=(urlon + lllon) / 2,
        lat_0=(urlat + lllat) / 2,
        llcrnrlat=lllat, urcrnrlat=urlat,
        llcrnrlon=lllon, urcrnrlon=urlon,
        resolution='f')
    # vẽ đường bờ biển, ranh giới tiểu bang và quốc gia, cạnh bản đồ.
    m.drawcoastlines()
    m.drawstates()
    m.drawcountries()
    return m
```

```python
# hiện thị số đồ thị với 2 hàng và 2 cột
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
fig.subplots_adjust(hspace=0.05, wspace=0.05)
# tạo các nhãn cho từng label
to_plot = ['2a', '1', '3c', '7a']
lllat=17.25; urlat=20.25; lllon=-75; urlon=-71 #setup các thông số đồ thị 

for code, ax in zip(to_plot, axes.flat):
    #  vẽ Basemap trên ô thích hợp, biến đổi tọa độ
    m = basic_haiti_map(ax, lllat=lllat, urlat=urlat,lllon=lllon, urlon=urlon)
    cat_data = data[data['category_%s' % code] == 1]
    # tính toán tọa độ bản đồ.
    x, y = m(cat_data.LONGITUDE, cat_data.LATITUDE)
    # hiện thị biểu đồ
    m.plot(x, y, 'k.', alpha=0.5)
    ax.set_title('%s: %s' % (code, english_mapping[code]))
```
